```python
df['Area'] = df['Area'].astype(float)
df['ConvexArea'] = df['ConvexArea'].astype(float)
```

## Scaling

```python
std_scaler = StandardScaler()
```

```python
features_to_scale = ['Area', 'Perimeter', 'MajorAxisLength', 'MinorAxisLength', 'ConvexArea', 'EquivDiameter']
for feature in features_to_scale:
    df[feature] = std_scaler.fit_transform(df[feature].values.reshape(-1, 1))
```

```python
df.head()
```

```python
df.describe()
```

## Categorical to Numerical

```python
df['Class'].unique()
```

```python
le_object = LabelEncoder()
```

```python
df['Class'] = le_object.fit_transform(df['Class'])
df.head()
```

## Seperating Dependent-Independent Variables

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
X
```

```python
y
```

## Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

## Saving Cleaned DFs to CSV files

```python
X_train.to_csv('./X_train.csv')
y_train.to_csv('./y_train.csv')
X_test.to_csv('./X_test.csv')
y_test.to_csv('./y_test.csv')
```

## Modelling using GridSearchCV

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [6, 8, 10],
    'min_samples_split': [20, 30, 40],
    'random_state': [28, 42]
}
```

```python
dtree_classifier = DecisionTreeClassifier()
```

```python
grid_search = GridSearchCV(estimator=dtree_classifier, param_grid=param_grid, cv=5, scoring='neg_log_loss')
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params_ = grid_search.best_params_
best_params_
```

```python
best_dtree_clf = DecisionTreeClassifier(**best_params_)
```

```python
best_dtree_clf.fit(X_train, y_train)
```

```python
y_pred = best_dtree_clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred, average = 'weighted')
```

```python
recall_score(y_test, y_pred, average = 'weighted')
```

```python
f1_score(y_test, y_pred, average = 'weighted')
```

```python
target_names = ['SEKER', 'BARBUNYA', 'BOMBAY', 'CALI', 'HOROZ', 'SIRA', 'DERMASON']
feature_names = ['Area', 'Perimeter',	'MajorAxisLength',	'MinorAxisLength',	'AspectRation',	'Eccentricity',	'ConvexArea',	'EquivDiameter',	'Extent',	'Solidity',	'roundness',	'Compactness',	'ShapeFactor1',	'ShapeFactor2',	'ShapeFactor3',	'ShapeFactor4']
```

```python
print(classification_report(y_true = y_test, y_pred = y_pred, target_names = target_names))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(
    best_dtree_clf,
    X_test,
    y_test,
    display_labels=target_names,
)
```
