## Scaling

```python
std_scaler = StandardScaler()
```

```python
features_to_scale = ['Area', 'Perimeter', 'MajorAxisLength', 'MinorAxisLength', 'ConvexArea', 'EquivDiameter']
for feature in features_to_scale:
    df[feature] = std_scaler.fit_transform(df[feature].values.reshape(-1, 1))
```

```python
df.head()
```

```python
df.describe()
```

## Categorical to Numerical

```python
df['Class'].unique()
```

```python
le_object = LabelEncoder()
```

```python
df['Class'] = le_object.fit_transform(df['Class'])
df.head()
```

## Seperating Dependent-Independent Variables

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
X
```

```python
y
```

## Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

## Saving Cleaned DFs to CSV files

```python
X_train.to_csv('./X_train.csv')
y_train.to_csv('./y_train.csv')
X_test.to_csv('./X_test.csv')
y_test.to_csv('./y_test.csv')
```

## Modelling

```python
ann_model = Sequential()
ann_model.add(Dense(units=64, input_dim=16, activation='relu'))
ann_model.add(Dense(units=7, activation='softmax'))
```

```python
ann_model.summary()
```

```python
ann_model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=['accuracy'])
```

```python
history = ann_model.fit(X_train,
        y_train,
        epochs = 100
        )
```

```python
ann_model.layers[0].get_weights()
```

```python
plt.plot(history.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
y_pred = ann_model.predict(X_test)
y_pred
```

```python
y_pred_labels = np.argmax(y_pred, axis=1)
y_pred_labels
```

```python
accuracy_score(y_test, y_pred_labels)
```

```python
precision_score(y_test, y_pred_labels, average = 'weighted')
```

```python
recall_score(y_test, y_pred_labels, average = 'weighted')
```

```python
f1_score(y_test, y_pred_labels, average = 'weighted')
```

```python
target_names = ['SEKER', 'BARBUNYA', 'BOMBAY', 'CALI', 'HOROZ', 'SIRA', 'DERMASON']
```

```python
print(classification_report(y_true = y_test, y_pred = y_pred_labels, target_names = target_names))
```

```python
confusion_matrix(y_test, y_pred_labels)
```
